```python
import matplotlib.pyplot as plt
```

```python
def tarefa(y0, r, n):
  y1 = r*y0*(1-y0)
  array = []
  x = []
  x.append(1)
  x.append(2)
  array.append(y0)
  array.append(y1)
  num = 3
  for i in range(n):
    yn = y1
    y = r*yn*(1-yn)
    array.append(y)
    x.append(num)
    num = num + 1
    y1 = y
  return array, num
```

```python
primeiro, x = tarefa(0.5, 0.5, 50)
print(primeiro)
```

```python
plt.plot(primeiro) 
```
